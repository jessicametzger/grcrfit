```python
import grcrfit.physics as ph
import grcrfit.crfluxes as crf
import numpy as np
import matplotlib.pyplot as plt

print(ph.LIS_DICT)
E=np.logspace(-1,2,num=50)

for key in ph.LIS_DICT:
    if key=='1h': continue
    plt.plot(E,crf.flux_spl_IS(ph.LIS_DICT[key], ph.E_to_p(E*1000., ph.M_DICT[key]), ph.M_DICT[key]),label=key)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
